```python
import matplotlib.pyplot as plt
```

```python
names = ["Ali", "Dana", "Murat", "Aruzhan", "Timur"]
scores = [85, 90, 78, 88, 92]
```

```python
plt.plot(names, scores)
plt.xlabel("Students")
plt.ylabel("Scores")
plt.title("Student Scores Line Plot")
plt.show()
```

```python
plt.bar(names, scores)
plt.xlabel("Students")
plt.ylabel("Scores")
plt.title("Student Scores Bar Chart")
plt.show()
```
