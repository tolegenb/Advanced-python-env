```python
import pandas as pd
```

```python
data = {
    "Name": ["Ali", "Dana", "Murat", "Aruzhan", "Timur"],
    "Age": [19, 22, 20, 23, 21],
    "Score": [85, 90, 78, 88, 92]
}

df = pd.DataFrame(data)
df
```

```python
print(df.head(3))
print(df.describe())
print("Средний балл:", df["Score"].mean())
```

```python
print("Старше 20 лет:")
print(df[df["Age"] > 20])

print("Score > 80:")
print(df[df["Score"] > 80])
```
