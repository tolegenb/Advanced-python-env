```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

print("Работает ✅")
```

```python
array_1d = np.arange(1, 11)
print("1D массив:", array_1d)

array_2d = np.random.randint(1, 10, size=(3, 3))
print("2D массив:\n", array_2d)
```

```python
print("Сумма:", array_1d.sum())
print("Среднее:", array_1d.mean())
print("Максимум:", array_1d.max())
```

```python
multiplied_array = array_1d * 2
print("Умножено на 2:", multiplied_array)
```

```python
modified_array = array_1d.copy()
modified_array[modified_array > 5] = 0
print("Значения > 5 заменены на 0:", modified_array)
```
